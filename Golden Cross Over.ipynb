```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta
```

```python
df = yf.download('TITAN.NS', start='2019-01-01')
```

```python
df['ma_200'] = df.Close.rolling(200).mean()
df['ma_50'] = df.Close.rolling(50).mean()
```

```python
df[['Close', 'ma_200', 'ma_50']].plot()
```

```python
df['rsi'] = ta.momentum.rsi(df.Close)
```

```python
df[['Close', 'rsi']].plot()
```

```python
df['rsi'].plot()
```

```python
df['vol'] = df.Close.rolling(20).std()
df['upper_bb'] = df.Close + (2*df.vol)
df['lower_bb'] = df.Close - (2*df.vol)
```

```python
df[['Close', 'upper_bb', 'lower_bb']].plot()
```
